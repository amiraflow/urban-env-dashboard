```python
def openaq_get(path: str, params: dict | None = None):
    """
    Small helper for GET requests to the OpenAQ v3 API.
    Raises an error if the response is not HTTP 200.
    """
    url = f"{BASE_URL.rstrip('/')}/{path.lstrip('/')}"
    response = requests.get(url, headers=HEADERS, params=params or {})
    
    try:
        data = response.json()
    except ValueError:
        # Not JSON
        response.raise_for_status()
        raise
    
    if not response.ok:
        # Show nice error from API if present
        raise RuntimeError(
            f"OpenAQ API error {response.status_code}: "
            f"{data.get('detail') or data}"
        )
    
    return data
```

```python
def get_localities_for_city(city_name, limit=500):
    """
    Returns all location records whose 'locality' contains the given city name.
    """
    data = openaq_get("locations", params={"limit": limit})
    df = pd.DataFrame(data.get("results", []))
    if "locality" not in df.columns:
        return pd.DataFrame()
    
    mask = df["locality"].fillna("").str.contains(city_name, case=False, na=False)
    return df[mask].reset_index(drop=True)
```

```python
def get_pm25_timeseries_for_location(location_id, limit=1000):
    """
    Fetch PM2.5 measurements for all sensors associated with a location.
    Returns a pandas DataFrame with timestamps and values.
    """
    sensors = openaq_get(f"locations/{location_id}/sensors", params={"limit": 100})
    sensors_df = pd.DataFrame(sensors.get("results", []))

    # Extract all PM2.5 sensors
    pm25_sensors = []
    for _, row in sensors_df.iterrows():
        params = row.get("parameters") or []
        if any(p.get("id") == 2 for p in params):
            pm25_sensors.append(row["id"])

    if not pm25_sensors:
        return pd.DataFrame()

    all_meas = []
    for sid in pm25_sensors:
        m = openaq_get(f"sensors/{sid}/measurements", params={"limit": limit})
        df = pd.DataFrame(m.get("results", []))
        df["sensor_id"] = sid
        all_meas.append(df)

    if not all_meas:
        return pd.DataFrame()

    df = pd.concat(all_meas, ignore_index=True)

    # Normalize datetime
    time_col = next((c for c in df.columns if "time" in c.lower() or "date" in c.lower()), None)
    if time_col:
        df["datetime"] = pd.to_datetime(df[time_col])
    else:
        return pd.DataFrame()

    return df[["datetime", "value", "sensor_id"]]  # keep only useful columns
```

```python
def get_all_locations(limit=100):
    all_results = []
    page = 1

    while True:
        data = openaq_get("locations", params={"limit": limit, "page": page})
        results = data.get("results", [])
        all_results.extend(results)

        meta = data.get("meta", {})
        if not meta.get("next"):
            break

        page += 1

    return pd.DataFrame(all_results)
```

```python
def get_all_locations(limit=1000, max_pages=50):
    """
    Fetch *all* locations across pages.

    limit: number of results per page (API max is usually 1000)
    max_pages: safety cap to avoid infinite loops
    """
    all_results = []
    page = 1

    while True:
        params = {"limit": limit, "page": page}
        data = openaq_get("locations", params=params)
        results = data.get("results", [])
        meta = data.get("meta", {}) or {}

        # Debug (you can comment this out once it works)
        print(f"Page {page}: got {len(results)} results, meta={meta}")

        if not results:
            break

        all_results.extend(results)

        # Try to use meta to know when to stop
        found_raw = meta.get("found")
        limit_raw = meta.get("limit", limit)
        page_raw = meta.get("page", page)

        try:
            found = int(found_raw) if found_raw is not None else None
            api_limit = int(limit_raw) if limit_raw is not None else limit
            current_page = int(page_raw) if page_raw is not None else page
        except (TypeError, ValueError):
            found = None
            api_limit = limit
            current_page = page

        stop_via_meta = (
            found is not None and current_page * api_limit >= found
        )

        # Fallback: if this page returned fewer than requested, we’re probably at the end
        stop_via_short_page = len(results) < limit

        if stop_via_meta or stop_via_short_page:
            break

        page += 1
        if max_pages is not None and page > max_pages:
            print("Reached max_pages limit, stopping pagination.")
            break

    return pd.DataFrame(all_results)
```

```python
def search_city(df, city_strings):
    """
    Search for city names across 'locality' and 'name' columns.
    city_strings can be a string or list of strings.
    """
    if isinstance(city_strings, str):
        city_strings = [city_strings]

    text = (
        df.get("locality", pd.Series(index=df.index, dtype="object")).fillna("")
        + " "
        + df.get("name", pd.Series(index=df.index, dtype="object")).fillna("")
    )

    mask = False
    for s in city_strings:
        mask = mask | text.str.contains(s, case=False, na=False)

    return df[mask]

vienna_locs = search_city(locations_df, ["Wien", "Vienna"])
sarajevo_locs = search_city(locations_df, ["Sarajevo"])

print("Vienna rows:", len(vienna_locs))
print("Sarajevo rows:", len(sarajevo_locs))

vienna_locs.head(), sarajevo_locs.head()
```

```python
def get_pm25_timeseries_for_location(location_id, limit=1000):
    sensors = openaq_get(f"locations/{location_id}/sensors", params={"limit": 1000})
    sensors_df = pd.DataFrame(sensors.get("results", []))

    if sensors_df.empty:
        return pd.DataFrame()

    pm25_sensors = []
    for _, row in sensors_df.iterrows():
        params = row.get("parameters") or []
        if any(p.get("id") == 2 for p in params):
            pm25_sensors.append(row["id"])

    if not pm25_sensors:
        return pd.DataFrame()

    all_meas = []
    for sid in pm25_sensors:
        m = openaq_get(f"sensors/{sid}/measurements", params={"limit": limit})
        df = pd.DataFrame(m.get("results", []))
        if df.empty:
            continue
        df["sensor_id"] = sid
        all_meas.append(df)

    if not all_meas:
        return pd.DataFrame()

    meas_df = pd.concat(all_meas, ignore_index=True)

    time_col = next(
        (c for c in meas_df.columns if "time" in c.lower() or "date" in c.lower()),
        None,
    )
    if time_col is None:
        return pd.DataFrame()

    meas_df["datetime"] = pd.to_datetime(meas_df[time_col])
    value_col = "value" if "value" in meas_df.columns else meas_df.columns[0]

    return meas_df[["datetime", value_col]].rename(columns={value_col: "value"})


def daily_mean(df):
    if df.empty:
        return df
    return (
        df.set_index("datetime")["value"]
        .sort_index()
        .resample("D")
        .mean()
    )


def city_pm25_daily(locality_df):
    if locality_df.empty:
        return None
    loc_id = locality_df.iloc[0]["id"]  # simple: use first location
    ts = get_pm25_timeseries_for_location(loc_id)
    if ts.empty:
        return None
    return daily_mean(ts)
```

```python
def get_pm25_series_for_city(loc_df, limit=5000):
    """
    For a city's set of locations:
      - gather all sensors measuring PM2.5 (parameter.id == 2)
      - fetch measurements from /sensors/{id}/measurements
      - return a merged DataFrame with columns ['datetime', 'value']
    """
    if loc_df.empty:
        return pd.DataFrame()

    pm25_sensor_ids = []

    # Collect PM2.5 sensor ids
    for _, row in loc_df.iterrows():
        sensors = row.get("sensors") or []
        for s in sensors:
            param = s.get("parameter", {}) or {}
            pid = param.get("id")
            if pid == 2:  # PM2.5
                pm25_sensor_ids.append(s["id"])

    pm25_sensor_ids = list(set(pm25_sensor_ids))  # unique
    if not pm25_sensor_ids:
        return pd.DataFrame()

    all_meas = []

    for sid in pm25_sensor_ids:
        try:
            data = openaq_get(f"sensors/{sid}/measurements", params={"limit": limit})
        except Exception as e:
            # skip problematic sensors quietly
            continue

        df = pd.DataFrame(data.get("results", []))
        if df.empty:
            continue

        # detect datetime column
        time_col = next(
            (c for c in df.columns if "time" in c.lower() or "date" in c.lower()),
            None
        )
        if not time_col:
            continue

        # detect value column
        value_col = "value" if "value" in df.columns else None
        if not value_col:
            continue

        df["datetime"] = pd.to_datetime(df[time_col], errors="coerce")
        df = df.dropna(subset=["datetime"])

        all_meas.append(df[["datetime", value_col]].rename(columns={value_col: "value"}))

    if not all_meas:
        return pd.DataFrame()

    merged = pd.concat(all_meas, ignore_index=True)
    merged = merged.sort_values("datetime")

    return merged
```

```python
def daily_mean(df):
    if df.empty:
        return df
    return (
        df.set_index("datetime")["value"]
        .sort_index()
        .resample("D")
        .mean()
        .dropna()
    )

city_daily_pm25 = {}

for city, locdf in city_locations.items():
    print(f"Building PM2.5 series for {city} ({len(locdf)} locations)...")
    ts = get_pm25_series_for_city(locdf)
    daily = daily_mean(ts)
    city_daily_pm25[city] = daily
    print(f"  -> {len(daily)} daily points")
```

```python
def get_pollutant_series_for_city(loc_df, pollutant_id, limit=5000):
    """
    Returns a merged timeseries (datetime → value) for a given pollutant id
    across all sensors belonging to all locations of a city.
    """
    if loc_df.empty:
        return pd.DataFrame()

    sensor_ids = []

    # Step 1: find all sensors measuring this pollutant
    for _, row in loc_df.iterrows():
        sensors = row.get("sensors") or []
        for s in sensors:
            param = s.get("parameter", {}) or {}
            pid = param.get("id")
            if pid == pollutant_id:
                sensor_ids.append(s["id"])

    if not sensor_ids:
        return pd.DataFrame()

    sensor_ids = list(set(sensor_ids))  # unique IDs

    # Step 2: download measurements from each sensor
    all_meas = []
    for sid in sensor_ids:
        try:
            data = openaq_get(f"sensors/{sid}/measurements", params={"limit": limit})
        except:
            continue

        df = pd.DataFrame(data.get("results", []))
        if df.empty:
            continue

        # autodetect datetime
        time_col = next(
            (c for c in df.columns if "time" in c.lower() or "date" in c.lower()),
            None
        )
        if not time_col:
            continue

        # autodetect value
        if "value" not in df.columns:
            continue

        df["datetime"] = pd.to_datetime(df[time_col], errors="coerce")
        df = df.dropna(subset=["datetime"])

        all_meas.append(df[["datetime", "value"]])

    if not all_meas:
        return pd.DataFrame()

    merged = pd.concat(all_meas, ignore_index=True)
    return merged.sort_values("datetime")
```

```python
def daily_mean(df):
    if df.empty:
        return df
    return (
        df.set_index("datetime")["value"]
        .sort_index()
        .resample("D")
        .mean()
        .dropna()
    )
```

```python
def sensor_daily_series(sensor_id: int, limit: int = 2000) -> pd.DataFrame:
    """
    Fetch daily-averaged measurements for a sensor using /days,
    return DataFrame with columns ['datetime', 'value'].
    """
    data = openaq_get(f"sensors/{sensor_id}/days", params={"limit": limit})
    df = pd.DataFrame(data.get("results", []))
    if df.empty:
        return df

    if "datetime" not in df.columns or "value" not in df.columns:
        return pd.DataFrame()

    # datetime is a dict with 'utc' and 'local' → pick 'utc'
    dt = df["datetime"].apply(
        lambda x: x.get("utc") if isinstance(x, dict) else x
    )
    df["datetime"] = pd.to_datetime(dt, errors="coerce")
    df = df.dropna(subset=["datetime"])

    return df[["datetime", "value"]].sort_values("datetime")
```

```python
POLLUTANTS = {
    "pm10": 1,
    "pm25": 2,
    "no2": 5,
    "o3": 7,
    "co": 4,
    "so2": 6,
}

def city_pollutant_daily(loc_df: pd.DataFrame, pollutant_id: int, limit_per_sensor: int = 2000) -> pd.Series:
    """
    For a city's locations:
      - find sensors whose parameter.id == pollutant_id
      - fetch /days for each
      - merge and average per day
    Returns a pandas Series indexed by datetime (day) with mean values.
    """
    if loc_df.empty:
        return pd.Series(dtype=float)

    sensor_ids = set()

    for _, row in loc_df.iterrows():
        for s in row.get("sensors") or []:
            param = s.get("parameter") or {}
            if param.get("id") == pollutant_id:
                sensor_ids.add(s["id"])

    if not sensor_ids:
        return pd.Series(dtype=float)

    all_daily = []

    for sid in sensor_ids:
        try:
            df = sensor_daily_series(sid, limit=limit_per_sensor)
        except Exception:
            continue
        if df.empty:
            continue
        all_daily.append(df)

    if not all_daily:
        return pd.Series(dtype=float)

    merged = pd.concat(all_daily, ignore_index=True)

    # aggregate across sensors per day (mean)
    merged["date"] = merged["datetime"].dt.date
    daily = merged.groupby("date")["value"].mean()

    # convert back to datetime index for plotting
    daily.index = pd.to_datetime(daily.index)
    daily = daily.sort_index()

    return daily
```

```python
def get_city_locations(df, city_strings, country_code=None):
    if isinstance(city_strings, str):
        city_strings = [city_strings]

    text = df["name"].fillna("") + " " + df["locality"].fillna("")

    mask = False
    for s in city_strings:
        mask |= text.str.contains(s, case=False, na=False)

    if country_code:
        mask &= df["country_code"].str.upper() == country_code.upper()

    return df[mask]
```

```python
def city_pollutant_daily_by_name(
    locdf,
    parameter_name: str,
    limit_per_sensor: int = 365,   # days per sensor
    max_sensors: int = 5           # cap to avoid too many requests
):
    """
    For a city's locations:
      - find sensors whose parameter.name == parameter_name
      - fetch /v3/sensors/{id}/days (daily aggregates)
      - take UTC start time from period.datetimeFrom
      - merge & average per day across sensors

    Returns a pandas Series indexed by date.
    """
    import pandas as pd

    if locdf.empty:
        return pd.Series(dtype=float)

    # 1) Collect sensor IDs for this pollutant
    sensor_ids = []
    for _, row in locdf.iterrows():
        for s in row.get("sensors") or []:
            param = s.get("parameter") or {}
            if param.get("name") == parameter_name:
                sid = s.get("id")
                if isinstance(sid, int):
                    sensor_ids.append(sid)

    sensor_ids = sorted(set(sensor_ids))
    if not sensor_ids:
        return pd.Series(dtype=float)

    # rate-limit friendly: use only first N sensors
    sensor_ids = sensor_ids[:max_sensors]

    all_daily = []

    for sid in sensor_ids:
        try:
            data = openaq_get(
                f"sensors/{sid}/days",
                params={"limit": min(limit_per_sensor, 1000)},  # API max
            )
        except RuntimeError as e:
            # e.g. 429 Too Many Requests → just skip this sensor
            print(f"Skipping sensor {sid}: {e}")
            continue

        df = pd.DataFrame(data.get("results", []))
        if df.empty or "value" not in df.columns or "period" not in df.columns:
            continue

        # Extract UTC start from period.datetimeFrom
        def extract_utc_start(period):
            if not isinstance(period, dict):
                return None
            dtobj = period.get("datetimeFrom") or {}
            if isinstance(dtobj, dict):
                return dtobj.get("utc") or dtobj.get("local")
            return None

        df["datetime"] = pd.to_datetime(
            df["period"].apply(extract_utc_start),
            errors="coerce",
        )
        df = df.dropna(subset=["datetime"])

        all_daily.append(df[["datetime", "value"]])

    if not all_daily:
        return pd.Series(dtype=float)

    merged = pd.concat(all_daily, ignore_index=True)

    # 3) Aggregate per calendar day across sensors
    merged["date"] = merged["datetime"].dt.date
    daily = merged.groupby("date")["value"].mean()
    daily.index = pd.to_datetime(daily.index)

    return daily.sort_index()
```

```python
def city_monthly_pollutant(locdf, parameter_name, max_sensors=10):
    """
    Build a monthly-mean pollutant series for all sensors in a city's location set.
    parameter_name: "pm10" or "pm25"
    """
    sensor_ids = set()

    # Collect sensor IDs for this pollutant
    for _, row in locdf.iterrows():
        for s in row.get("sensors") or []:
            p = s.get("parameter")
            if p and p.get("name") == parameter_name:
                sensor_ids.add(s["id"])

    # Limit sensors to avoid too many API calls
    sensor_ids = list(sensor_ids)[:max_sensors]

    if not sensor_ids:
        return pd.Series(dtype=float)

    all_data = []
    for sid in sensor_ids:
        data = openaq_get(f"sensors/{sid}/days", params={"limit": 1000})
        df = pd.DataFrame(data.get("results", []))

        if df.empty or "datetime" not in df or "value" not in df:
            continue

        df["datetime"] = pd.to_datetime(
            df["datetime"].apply(lambda d: d.get("utc") if isinstance(d, dict) else d),
            errors="coerce",
            utc=True
        )

        df = df.dropna(subset=["datetime"])
        df = df[["datetime", "value"]]
        all_data.append(df)

    if not all_data:
        return pd.Series(dtype=float)

    merged = pd.concat(all_data, ignore_index=True)
    merged["date"] = merged["datetime"].dt.to_period("M").dt.to_timestamp()
    monthly = merged.groupby("date")["value"].mean()
    return monthly.sort_index()
```

```python
def city_monthly_pollutant(locdf, parameter_name, max_sensors=5):
    """
    Build a monthly-mean pollutant series for all sensors in a city's locations.
    parameter_name: "pm10" or "pm25"
    """
    import pandas as pd

    if locdf.empty:
        return pd.Series(dtype=float)

    # 1) Collect sensor IDs with the desired pollutant
    sensor_ids = []
    for _, row in locdf.iterrows():
        for s in row.get("sensors") or []:
            p = s.get("parameter") or {}
            if p.get("name") == parameter_name:
                sid = s.get("id")
                if isinstance(sid, int):
                    sensor_ids.append(sid)

    sensor_ids = list(dict.fromkeys(sensor_ids))  # unique, preserve order
    sensor_ids = sensor_ids[:max_sensors]         # rate limit

    if not sensor_ids:
        return pd.Series(dtype=float)

    all_data = []

    for sid in sensor_ids:
        try:
            data = openaq_get(f"sensors/{sid}/days", params={"limit": 1000})
        except Exception as e:
            print(f"Skipping sensor {sid}: {e}")
            continue

        df = pd.DataFrame(data.get("results", []))
        if df.empty or "value" not in df.columns or "period" not in df.columns:
            continue

        # Extract UTC start time from period.datetimeFrom
        def extract_utc_start(period):
            if not isinstance(period, dict):
                return None
            dtobj = period.get("datetimeFrom") or {}
            if isinstance(dtobj, dict):
                return dtobj.get("utc") or dtobj.get("local")
            return None

        df["datetime"] = pd.to_datetime(
            df["period"].apply(extract_utc_start),
            errors="coerce",
            utc=True,
        )
        df = df.dropna(subset=["datetime"])

        all_data.append(df[["datetime", "value"]])

    if not all_data:
        return pd.Series(dtype=float)

    merged = pd.concat(all_data, ignore_index=True)

    # 2) Monthly mean (calendar months)
    merged = merged.set_index("datetime")
    monthly = merged["value"].resample("ME").mean()

    return monthly.sort_index()
```

```python
def monthly_coverage(series):
    """Calculate % of days in each month that have valid readings."""
    if series is None or series.empty:
        return pd.Series(dtype=float)

    # Count number of valid days each month
    daily_count = series.resample("D").count()
    monthly_days = daily_count.resample("M").count()
    
    # How many days were actually present?
    reported_days = daily_count.resample("M").sum()
    
    # % coverage
    coverage = (reported_days / monthly_days) * 100
    return coverage
```

```python
def extract_openaq_datetime(row):
    """
    Extract a UTC timestamp from all possible OpenAQ formats:
    - row["datetime"]["utc"]
    - row["period"]["datetimeFrom"]["utc"]
    - row["local"] etc. (fallback)
    Returns a pandas.Timestamp or NaT.
    """
    # Format 1: row["datetime"]["utc"]
    if isinstance(row.get("datetime"), dict):
        utc = row["datetime"].get("utc")
        if utc:
            return pd.to_datetime(utc, errors="coerce")

    # Format 2: row["period"]["datetimeFrom"]["utc"]
    if isinstance(row.get("period"), dict):
        start = row["period"].get("datetimeFrom")
        if isinstance(start, dict):
            utc = start.get("utc")
            if utc:
                return pd.to_datetime(utc, errors="coerce")

    # Nothing found → NaT
    return pd.NaT



def city_daily_pollutant(city_df, parameter_name, limit_per_sensor=1000):
    """Return a reliable daily pollutant series for the city."""
    sensor_ids = set()

    # Collect all sensors for this pollutant
    for _, row in city_df.iterrows():
        for s in row.get("sensors") or []:
            p = s.get("parameter") or {}
            if p.get("name") == parameter_name:
                sensor_ids.add(s["id"])

    all_daily = []

    # Fetch daily data from each sensor
    for sid in sensor_ids:
        try:
            data = openaq_get(f"sensors/{sid}/days", params={"limit": min(limit_per_sensor, 1000)})
        except Exception:
            continue

        results = data.get("results", [])
        if not results:
            continue

        df = pd.DataFrame(results)

        # Extract timestamps robustly
        df["timestamp"] = df.apply(extract_openaq_datetime, axis=1)

        # Must have timestamp and value
        if "value" not in df.columns:
            continue

        df = df.dropna(subset=["timestamp", "value"])
        if df.empty:
            continue

        df = df.set_index("timestamp")
        all_daily.append(df["value"])

    if not all_daily:
        # no data found
        return pd.Series(dtype=float)

    merged = pd.concat(all_daily)
    # Combine multiple sensors → daily mean
    daily = merged.groupby(merged.index.date).mean()
    daily.index = pd.to_datetime(daily.index)

    return daily.sort_index()
```

```python
def monthly_coverage_from_daily(daily_series):
    if daily_series is None or daily_series.empty:
        return pd.Series(dtype=float)

    # Count daily values
    days_present = daily_series.resample("D").count()

    # Compute: how many days exist per month?
    reported = days_present.resample("ME").sum()
    total = days_present.resample("ME").count()

    return (reported / total * 100).fillna(0)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------------------------------
# 1. Helper: fuzzy city/location matching in locations_df
# -------------------------------------------------------
def get_city_locations(df, city_strings, country_code=None):
    """
    Return subset of locations_df for a given city (or city-like region),
    matching any of the given strings in name or locality, case-insensitive.
    """
    if isinstance(city_strings, str):
        city_strings = [city_strings]

    name_locality = (
        df["name"].fillna("").astype(str) + " " +
        df["locality"].fillna("").astype(str)
    )

    mask = False
    for s in city_strings:
        mask |= name_locality.str.contains(s, case=False, na=False, regex=False)

    if country_code:
        mask &= df["country_code"].str.upper().eq(country_code.upper())

    return df[mask]


# -------------------------------------------------------
# 2. Helper: robust datetime extraction from OpenAQ /days
# -------------------------------------------------------
def extract_openaq_datetime(row):
    """
    Extract a UTC timestamp from all possible OpenAQ v3 daily formats:
    - row["datetime"]["utc"]
    - row["period"]["datetimeFrom"]["utc"]
    Returns pandas.Timestamp or NaT.
    """
    # Format 1
    dt = row.get("datetime")
    if isinstance(dt, dict):
        utc = dt.get("utc")
        if utc:
            return pd.to_datetime(utc, errors="coerce", utc=True)

    # Format 2
    period = row.get("period")
    if isinstance(period, dict):
        start = period.get("datetimeFrom")
        if isinstance(start, dict):
            utc = start.get("utc")
            if utc:
                return pd.to_datetime(utc, errors="coerce", utc=True)

    return pd.NaT


# -------------------------------------------------------
# 3. Helper: build monthly mean PM time series for a city
# -------------------------------------------------------
def city_monthly_pollutant(city_df, parameter_name, max_sensors=5, limit_per_sensor=1000):
    """
    For a city's locations:
      - find all sensors whose parameter.name == parameter_name (e.g. "pm10")
      - call /v3/sensors/{id}/days for each sensor
      - extract daily values and aggregate to monthly means

    Returns:
      pandas.Series indexed by month-end (Timestamp), with monthly mean values.
    """
    if city_df.empty:
        return pd.Series(dtype=float)

    # Collect sensor IDs for this pollutant
    sensor_ids = []
    for _, row in city_df.iterrows():
        for s in row.get("sensors") or []:
            p = s.get("parameter") or {}
            if p.get("name") == parameter_name:
                sid = s.get("id")
                if isinstance(sid, int):
                    sensor_ids.append(sid)

    # Unique + rate limit
    sensor_ids = list(dict.fromkeys(sensor_ids))[:max_sensors]

    if not sensor_ids:
        print(f"[WARN] No sensors for {parameter_name} in this city selection.")
        return pd.Series(dtype=float)

    all_daily = []

    for sid in sensor_ids:
        try:
            data = openaq_get(f"sensors/{sid}/days", params={"limit": min(limit_per_sensor, 1000)})
        except Exception as e:
            print(f"[WARN] Skipping sensor {sid} – API error: {e}")
            continue

        results = data.get("results", [])
        if not results:
            continue

        df = pd.DataFrame(results)
        if "value" not in df.columns:
            continue

        # Extract timestamp
        df["timestamp"] = df.apply(extract_openaq_datetime, axis=1)
        df = df.dropna(subset=["timestamp", "value"])
        if df.empty:
            continue

        df = df.set_index("timestamp")
        all_daily.append(df["value"])

    if not all_daily:
        print(f"[WARN] No daily data for {parameter_name} after fetching sensors.")
        return pd.Series(dtype=float)

    # Combine all sensors & average per day
    merged = pd.concat(all_daily)
    daily = merged.groupby(merged.index.date).mean()
    daily.index = pd.to_datetime(daily.index, utc=True)
    daily = daily.sort_index()

    # Monthly-end means (ME instead of deprecated M)
    monthly = daily.resample("ME").mean()
    return monthly


# -------------------------------------------------------
# 4. Build city location DataFrames for EU cities
# -------------------------------------------------------

vienna_df   = get_city_locations(locations_df, ["Wien", "Vienna"], "AT")
zagreb_df   = get_city_locations(locations_df, ["Zagreb"], "HR")
belgrade_df = get_city_locations(locations_df, ["Beograd", "Belgrade"], "RS")
berlin_df   = get_city_locations(locations_df, ["Berlin"], "DE")
madrid_df   = get_city_locations(locations_df, ["Madrid", "MADRID"], "ES")
stockholm_df = get_city_locations(locations_df, ["Stockholm"], "SE")
praha_df = get_city_locations(locations_df, ["Praha"], "CZ")
sarajevo_df = get_city_locations(locations_df, ["Sarajevo"], "BA")

print("Vienna locations:", len(vienna_df))
print("Zagreb locations:", len(zagreb_df))
print("Belgrade locations:", len(belgrade_df))
print("Berlin locations:", len(berlin_df))
print("Madrid locations:", len(madrid_df))
print("Stockholm locations:", len(stockholm_df))
print("Praha locations:", len(praha_df))
print("Sarajevo locations:", len(sarajevo_df))

cities_eu_df = {
    "Vienna": vienna_df,
    "Zagreb": zagreb_df,
    "Belgrade": belgrade_df,
    "Berlin": berlin_df,
    "Madrid": madrid_df,
    "Stockholm": stockholm_df,
    #"Praha": praha_df,
    #"Sarajevo": sarajevo_df,
}


# -------------------------------------------------------
# 5. Compute monthly PM10 for each EU city
# -------------------------------------------------------
monthly_pm10_eu = {}

for cname, cdf in cities_eu_df.items():
    print(f"\nProcessing PM10 for {cname} (locations: {len(cdf)})...")
    series = city_monthly_pollutant(cdf, "pm10", max_sensors=5, limit_per_sensor=1000)
    print(f"  -> {len(series)} monthly points")
    monthly_pm10_eu[cname] = series


# -------------------------------------------------------
# 6. Plot: EU PM10 comparison for 2019–2022
# -------------------------------------------------------
plot_start = pd.Timestamp("2019-01-01", tz="UTC")
plot_end   = pd.Timestamp("2022-12-31", tz="UTC")

eu_pm10_trimmed = {}

for cname, s in monthly_pm10_eu.items():
    if s is None or s.empty:
        continue
    trimmed = s[(s.index >= plot_start) & (s.index <= plot_end)]
    eu_pm10_trimmed[cname] = trimmed
    print(f"{cname}: {len(trimmed)} monthly points in {plot_start.date()} – {plot_end.date()}")

plt.figure(figsize=(14, 6))

for cname, s in eu_pm10_trimmed.items():
    if not s.empty:
        plt.plot(s.index, s.values, label=cname, linewidth=2)

plt.title("Monthly PM10 Levels Across European Cities (2019–2022)", fontsize=14)
plt.xlabel("Date")
plt.ylabel("PM10 (µg/m³)")
plt.grid(alpha=0.3, linestyle="--")
plt.legend(ncol=3, frameon=False)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# -------------------------------------------------------
# 1. Load OWID PM2.5 dataset
#    Source: Our World in Data - "Average annual exposure to PM2.5 air pollution"
#    https://ourworldindata.org/grapher/average-exposure-pm25-pollution
# -------------------------------------------------------

owid_url = "https://ourworldindata.org/grapher/average-exposure-pm25-pollution.csv"
owid_pm25 = pd.read_csv(owid_url)

# Inspect the columns (run once if you want to see structure)
print(owid_pm25.head())
print(owid_pm25.columns)

# Typical columns for OWID grapher:
# 'Entity', 'Code', 'Year', 'average_exposure_pm25_polution' (or similar)
# Let's automatically detect the value column as "the last one"
value_col = owid_pm25.columns[-1]

# -------------------------------------------------------
# 2. Filter to a single recent year (e.g. 2020)
# -------------------------------------------------------

target_year = 2020
df_2020 = owid_pm25[owid_pm25["Year"] == target_year].copy()

# Drop aggregates / rows with no country code
df_2020 = df_2020.dropna(subset=["Code"])

# Rename columns for convenience
df_2020 = df_2020.rename(columns={
    "Entity": "country",
    "Code": "iso_code",
    value_col: "pm25"
})

print(f"Number of countries in {target_year}:", len(df_2020))

# -------------------------------------------------------
# 3. Plot choropleth map of PM2.5 exposure
# -------------------------------------------------------

fig = px.choropleth(
    df_2020,
    locations="iso_code",
    color="pm25",
    hover_name="country",
    color_continuous_scale="Viridis",
    range_color=(0, df_2020["pm25"].quantile(0.99)),  # avoid extreme outliers
    title=f"Average Annual Exposure to PM2.5 Air Pollution ({target_year})",
)

fig.update_layout(
    coloraxis_colorbar=dict(
        title="PM2.5 (µg/m³)"
    ),
    margin=dict(l=0, r=0, t=60, b=0)
)

fig.show()
```
